```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
chunk_sizes = ['full_chunks', 'half_chunks', 'sqrt_chunks', 'single_chunks', 'no_chunks']
speedups_df = pd.DataFrame(columns=chunk_sizes, index=range(1,8))

for chunk_size in chunk_sizes:
    df = pd.read_csv('bench_results_%s.csv' % chunk_size, index_col=1)
    df.drop('Unnamed: 0', axis=1, inplace=True)

    means = df.mean(axis=0)    
    speedups_df[chunk_size] = list(means.iloc[0] / means)
```

```python
speedups_df.plot()
plt.plot(range(1,8), range(1,8), 'r')
```
